


---

Los valores faltantes identificados en las columnas categóricas 'job', 'marital', 'education', 'contact', y 'poutcome' serán imputados utilizando la moda, es decir, el valor más frecuente dentro de cada columna. Dado que estas variables son categóricas, la aplicación de métodos de imputación basados en medidas de tendencia central numéricas, como la media o la mediana, no es coherente. La imputación por la moda, en cambio, es un método que garantiza que los valores imputados pertenezcan al conjunto de categorías ya existentes en el dataset, evitando la introducción de valores inválidos. Además, al reemplazar los valores faltantes con la categoría más común, se preserva la frecuencia relativa de dicha categoría, lo cual puede ser una consideración importante si esa clase es representativa de una parte significativa de las observaciones.

---

```python
categorical_cols_to_impute = ['job', 'marital', 'education', 'contact', 'poutcome']

print("\nImputando missing values con la moda para las columnas categóricas:")
for col in categorical_cols_to_impute:
    if df[col].isnull().any(): # Verificar si todavía hay nulos en caso de ejecuciones previas parciales
        mode_value = df[col].mode()[0] # Calcular la moda (mode() devuelve una Serie, tomamos el primer elemento)
        df[col].fillna(mode_value, inplace=True)
        print(f"Imputados missing values en '{col}' con la moda: '{mode_value}'")
```

### Pregunta 1.4

 Cree un _pipeline_ de preprocesamiento con las siguientes transformaciones a las columnas correspondientes:

*   **Columnas Numéricas** : Aplique un `StandardScaler` para estandarizar estas variables.
*   **Columnas Categóricas Binarias** : Aplique `OneHotEncoder` para crear variables dummy. Asegúrese de manejar posibles categorías desconocidas durante la transformación.
*   **Columnas a Eliminar** : Elimine columnas que dejan de ser relavantes luego de las trasformaciones.
*  **Columnas Dummies MultiNivel** : cree columnas dummies para las variables multi categoricas (tomando codificadas a 0s y 1s =, para su respctiva categoria)
*  **Columnas Label Encoding MultiNivel** : Dentro de las columnas multinivel hay 2 columnas que presentan un orden natural. e.g. en la columna 'month' tenemos que 'jan' < 'feb' < 'dec'. Aplique en la columna 'month' y otra columna que deberá determinar usted un label encoder que capture ese ordenamiento natural, luego normalice con un minmax scaler para que las columnas varien entre 0 y 1.


Una vez creado el pipeline, aplíquelo a su conjunto de features (`X`). Muestre la forma (_shape_) del dataset original y del dataset procesado para verificar los cambios. ¿Cuántas nuevas columnas hay?

Adicionalmente, elimine del análisis la variable `day_of_week`. Si considera necesario realizar transformaciones adicionales a las variables, realicelas en este punto.



```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline

# Assuming df has been created from concatenating X and y
# Ensure categorical columns have no missing values by imputing with the mode
categorical_cols_with_potential_nan = ['job', 'marital', 'education', 'contact', 'poutcome']
for col in categorical_cols_with_potential_nan:
    if df[col].isnull().any():
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)
        print(f"Imputed missing values in '{col}' with mode: '{mode_value}'")

# Now, extract features (X_imputed) and target (y_imputed) from the cleaned df
X_imputed = df[X.columns]
y_imputed = df[y.columns]

# Identificar tipos de columnas
numerical_features = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']
binary_categorical_features = ['default', 'housing', 'loan']
multilevel_categorical_features = ['job', 'marital', 'contact', 'poutcome']
ordered_categorical_features = ['education', 'month']
columns_to_drop = ['day_of_week']

# Definir el orden para las columnas categóricas ordenadas
education_order = ['unknown', 'primary', 'secondary', 'tertiary']
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

# Crear el ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('binary_cat', OneHotEncoder(handle_unknown='ignore'), binary_categorical_features),
        ('multilevel_cat_ohe', OneHotEncoder(handle_unknown='ignore'), multilevel_categorical_features),
        ('ordered_cat_label', Pipeline([
            ('ordinal', OrdinalEncoder(categories=[education_order, month_order])),
            ('minmax', MinMaxScaler())
        ]), ordered_categorical_features),
        ('drop_cols', 'drop', columns_to_drop)
    ],
    remainder='passthrough'
)

# Aplicar el preprocesamiento to the imputed data
X_processed = preprocessor.fit_transform(X_imputed)

# Mostrar la forma del dataset original y procesado
print(f"Forma del dataset original (X): {X.shape}")
print(f"Forma del dataset procesado (X_processed): {X_processed.shape}")

# Calcular el número de nuevas columnas
original_cols = X.shape[1]
processed_cols = X_processed.shape[1]
print(f"Número de columnas eliminadas: {len(columns_to_drop)}")
print(f"Número total de nuevas columnas generadas: {processed_cols - (original_cols - len(columns_to_drop))}")
print(f"Número total de columnas en el dataset procesado: {processed_cols}")
print(f"Diferencia total de columnas: {processed_cols - original_cols}")
```

En caso de que el conjunto de datos no esté balanceado, asegúrese de que la división de _train_-_test_ respete el balance de las clases target para evitar que se genere un sesgo. Para esto, se recomienda leer la documentación oficial de la función `train_test_split()`: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html.

Use la semilla `random_state` = 123.


```python
from sklearn.model_selection import train_test_split

# Asumiendo que X_processed e y_imputed están disponibles de los pasos anteriores

# Verificar la distribución de la variable objetivo
print("Distribución de la variable objetivo en el dataset completo:")
print(y_imputed.value_counts(normalize=True))

X_train, X_test, y_train, y_test = train_test_split(
    X_processed,
    y_imputed,
    test_size=0.25,
    random_state=123,
    stratify=y_imputed
)

print("\nDistribución de la variable objetivo en el conjunto de entrenamiento:")

if isinstance(y_train, np.ndarray):
    y_train_series = pd.Series(y_train.flatten())
else:
    y_train_series = y_train.iloc[:, 0]
print(y_train_series.value_counts(normalize=True))


print("\nDistribución de la variable objetivo en el conjunto de prueba:")

if isinstance(y_test, np.ndarray):
    y_test_series = pd.Series(y_test.flatten())
else:
    y_test_series = y_test.iloc[:, 0]
print(y_test_series.value_counts(normalize=True))


is_imbalanced = y_imputed.value_counts(normalize=True).min() < 0.4 # Se considera como balanceado si es sobre el umbral de 0.4
if is_imbalanced:
    print("\nEl conjunto de datos está desbalanceado.")
    print("La división train-test se realizó utilizando stratify para mantener el balance de clases.")
else:
    print("\nEl conjunto de datos parece estar balanceado.")
```

### Pregunta 1.6

Entrene un árbol clasificador sin definir profundad máxima del árbol. Llamaremos a este, "modelo no regularizado".

Asegúrese de que el modelo sea capaz de lidiar con potenciales desbalances. Para esto se recomienda leer la documentación oficial de la *clase* `DecisionTreeClassifier()` (parámetros de inicialización): https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html.


Despliege el $F_1\ Score$ del modelo sobre la base de prueba. Luego, grafique una representación del árbol entrenado. ¿Qué puede decir sobre este modelo no regularizado?

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import f1_score


modelo_dt_no_regularizado = DecisionTreeClassifier(random_state=123, class_weight='balanced')

# Entrenar el modelo con los datos de entrenamiento
modelo_dt_no_regularizado.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_no_regularizado = modelo_dt_no_regularizado.predict(X_test)

# Calcular el F1-Score en el conjunto de prueba
f1_no_regularizado = f1_score(y_test, y_pred_no_regularizado, pos_label='yes') # Asumiendo 'yes' es la clase positiva

print(f"F1-Score del modelo de árbol de decisión no regularizado en la base de prueba: {f1_no_regularizado:.4f}")

# Graficar el árbol entrenado
# Debido a que el árbol no está regularizado, puede ser muy grande.
# Limitar la profundidad para la visualización si es necesario.
plt.figure(figsize=(20, 10))
plot_tree(modelo_dt_no_regularizado, feature_names=preprocessor.get_feature_names_out(X.columns),
          class_names=modelo_dt_no_regularizado.classes_, filled=True)
plt.title("Árbol de Decisión No Regularizado (visualización ilimitada)")
plt.show()
```




---

Al entrenar un Árbol de Decisión sin imponer restricciones en su profundidad máxima, permitimos que el árbol crezca hasta que cada nodo hoja sea puro o contenga un número mínimo de muestras. Este enfoque resulta en un modelo "no regularizado". La principal característica de un árbol no regularizado es su tendencia a ajustarse de manera extremadamente cercana a los datos de entrenamiento, llegando incluso a capturar el ruido presente en ellos.

Este ajuste excesivo a los datos de entrenamiento se conoce como *sobreajuste* (overfitting). El árbol se vuelve hipersensible a los patrones específicos de la muestra de entrenamiento, lo que limita su capacidad para generalizar bien a datos nuevos y no vistos, como los de la base de prueba. En consecuencia, su rendimiento en la base de prueba suele no ser bueno, reflejando su incapacidad para manejar variaciones fuera del conjunto de entrenamiento.

Visualizar un árbol no regularizado en su totalidad a menudo revela una estructura extremadamente compleja y profunda, con numerosas ramas y nodos. Esta complejidad inherente, si bien le permite memorizar el conjunto de entrenamiento, dificulta su interpretabilidad. Resulta dificíl seguir las reglas de decisión del árbol para entender por qué una observación particular es clasificada de cierta manera. La falta de interpretabilidad es una desventaja significativa, especialmente en contextos donde es crucial entender los factores que impulsan una predicción.

Entrene un árbol clasificador con profundad máxima del árbol igual a 16. Llamaremos a este, "modelo levemente regularizado". Asegúrese de que el modelo sea capaz de lidiar con potenciales desbalances. Luego, despliege el $F_1\ Score$ del modelo sobre la base de prueba.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt


modelo_dt_levemente_regularizado = DecisionTreeClassifier(max_depth=16, random_state=123, class_weight='balanced')

# Entrenar el modelo con los datos de entrenamiento
modelo_dt_levemente_regularizado.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_levemente_regularizado = modelo_dt_levemente_regularizado.predict(X_test)

# Calcular el F1-Score en el conjunto de prueba
f1_levemente_regularizado = f1_score(y_test, y_pred_levemente_regularizado, pos_label='yes') # Asumiendo 'yes' es la clase positiva

print(f"F1-Score del modelo de árbol de decisión levemente regularizado en la base de prueba: {f1_levemente_regularizado:.4f}")

# Graficar el árbol entrenado
# Limitar la profundidad para la visualización si es necesario, ya que 16 niveles pueden ser demasiados para mostrar claramente.
plt.figure(figsize=(20, 10))
plot_tree(modelo_dt_levemente_regularizado, max_depth=16, feature_names=preprocessor.get_feature_names_out(X.columns),
          class_names=modelo_dt_levemente_regularizado.classes_, filled=True)
plt.title("Árbol de Decisión Levemente Regularizado (visualización limitada a profundidad 16)")
plt.show()
```



---

Al limitar la profundidad máxima del árbol a 16, estamos aplicando una forma de regularización. Esta restricción controla el crecimiento del árbol, haciéndolo menos complejo que un árbol no regularizado (con profundidad ilimitada).

El objetivo de esta regularización es mitigar el sobreajuste que ocurre cuando el modelo se ajusta demasiado a los detalles específicos de los datos de entrenamiento, comprometiendo su capacidad para generalizar a datos nuevos y no vistos. Mientras que el árbol no regularizado intenta alcanzar una pureza casi total en las hojas memorizando el conjunto de entrenamiento (lo que probablemente resultaría en un F1-Score de entrenamiento muy alto pero un F1-Score de prueba menor debido al sobreajuste), al establecer max_depth=8, forzamos al árbol a encontrar reglas de decisión más generales.

Al reducir la complejidad del modelo de esta manera, esperamos que capture los patrones más importantes y generales de los datos, en lugar de memorizar la muestra de entrenamiento. Esto se traduce en una mejor capacidad de generalización. El F1-Score calculado en la base de prueba es una métrica clave para evaluar esta generalización. Es por esto que el F1-Score en la base de prueba mejora en comparación con el modelo no regularizado porque la limitación de profundidad ayuda a reducir el sobreajuste y a encontrar un mejor equilibrio entre el sesgo (error por simplificar demasiado) y la varianza (sensibilidad al ruido).


En conclusión, el modelo de árbol de decisión regularizado con max_depth=16 es un intento de mejorar la capacidad de generalización del modelo reduciendo el sobreajuste. Basándonos en los principios de los árboles de decisión y la regularización, anticipamos que el F1-Score de este modelo en la base de prueba será superior al del árbol no regularizado

### Pregunta 1.10

Entrene 20 árboles de clasificación. Cada uno de ellos debe estar entrenado fijando la profundidad máxima del árbol `max_depth` igual a $n \in [1,20]$. Obtenga el $F_1\ Score$ de cada árbol con la base de entrenamiento y la base de prueba.

Gafique en una misma figura $F_1\ Score$ de entrenamiento contra $n$ y $F_1\ Score$ de prueba contra $n$ (Eje X = $n$). ¿Cuál parece ser el parámetro óptimo $n^*$ para el árbol? ¿Qué puede decir sobre el parámetro sobre _overfitting_ y generalización?

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

# Listas para almacenar los F1-Scores de entrenamiento y prueba
f1_train_scores = []
f1_test_scores = []

# Rango de profundidades a evaluar
max_depth_range = range(1, 21)

# Entrenar 20 árboles con diferentes profundidades
for n in max_depth_range:
    # Inicializar el modelo de árbol de decisión con la profundidad actual
    # Usar class_weight='balanced' para manejar el desbalance
    modelo_dt = DecisionTreeClassifier(max_depth=n, random_state=123, class_weight='balanced')

    # Entrenar el modelo con los datos de entrenamiento
    modelo_dt.fit(X_train, y_train)

    # Realizar predicciones en los conjuntos de entrenamiento y prueba
    y_train_pred = modelo_dt.predict(X_train)
    y_test_pred = modelo_dt.predict(X_test)

    # Calcular el F1-Score para entrenamiento y prueba
    # Asegúrate de que 'yes' es la etiqueta correcta para la clase positiva
    f1_train = f1_score(y_train, y_train_pred, pos_label='yes')
    f1_test = f1_score(y_test, y_test_pred, pos_label='yes')

    # Almacenar los scores
    f1_train_scores.append(f1_train)
    f1_test_scores.append(f1_test)

# Graficar los F1-Scores
plt.figure(figsize=(12, 6))
plt.plot(max_depth_range, f1_train_scores, marker='o', label='F1-Score de Entrenamiento')
plt.plot(max_depth_range, f1_test_scores, marker='o', label='F1-Score de Prueba')
plt.title('F1-Score vs. Profundidad Máxima del Árbol')
plt.xlabel('Profundidad Máxima del Árbol (n)')
plt.ylabel('F1-Score')
plt.xticks(max_depth_range)
plt.grid(True)
plt.legend()
plt.show()

# Encontrar el parámetro óptimo n* (el que maximiza el F1-Score en la base de prueba)
optimal_n_index = f1_test_scores.index(max(f1_test_scores))
optimal_n = max_depth_range[optimal_n_index]
optimal_f1_test = f1_test_scores[optimal_n_index]

print(f"El parámetro óptimo n* (que maximiza el F1-Score en la base de prueba) parece ser: {optimal_n}")
print(f"El F1-Score correspondiente en la base de prueba es: {optimal_f1_test:.4f}")
```



---

La profundidad máxima del árbol (n) juega un papel crucial en el equilibrio entre sobreajuste y generalización. Al aumentar el valor de n, el árbol se vuelve más complejo, lo que le permite ajustarse más intrincadamente a los datos de entrenamiento. Esto se refleja en un F1-Score de entrenamiento que generalmente aumenta o se mantiene en niveles altos a medida que la profundidad aumenta. Sin embargo, esta mayor complejidad no siempre se traduce en un mejor rendimiento en datos no vistos. El F1-Score en la base de prueba tiende a aumentar inicialmente con n, ya que el modelo aprende patrones más complejos que son útiles para la predicción. No obstante, si la profundidad excede un cierto punto, el modelo comienza a memorizar el ruido y las particularidades específicas del conjunto de entrenamiento, lo que deteriora su capacidad para generalizar a datos nuevos. Este fenómeno es el sobreajuste, y se observa cuando el F1-Score de entrenamiento sigue siendo alto o aumenta, mientras que el F1-Score de prueba disminuye. Para este análisis, el parámetro óptimo n* parece ser 7, logrando el mejor F1-Score en la base de prueba. Este punto marca el mejor equilibrio del modelo entre ajustarse a los datos de entrenamiento y generalizar a datos no vistos, evitando el sobreajuste que ocurriría con profundidades mayores.

## Missing Values

### Pregunta 2.4

Realice análisis de valores vacíos y tratamiento de estos. Se premiará mantener el mayor número de observaciones posibles bajo criterios razonables. Luego de esta pregunta no deben quedar nulos en el _dataframe_ a utilizar.

Lo primero que vamos a hacer es lo siguiente: inputaremos datos faltantes como el promedio de los países vecinos. La idea es poder llenar los NA con valores que sean ojalá lo más fidedignos posibles pero también no perjudicar la naturaleza de la varianza para mantener la mayor información, cosa que no se lograría si inputamos todos los faltantes con la media o la mediana.

```python
url = "https://raw.githubusercontent.com/raul12101/Tarea-2-Machine-Learning-/refs/heads/main/country_neighbours.json"

# Obtenemos el diccionario creado en el otro notebook (country_neighbours.ipynb).
# Se decidió hacerlo así debido a que geopandas toma tiempo instalar, y también tarda unos segundos
# en obtener el diccionario.
response = requests.get(url)
neighbours_dict = response.json()

print(neighbours_dict)
```

Realizamos el ajuste a la media de los países vecinos, de existir:

```python
for key_country in neighbours_dict:
    if key_country in df["Country Name"].values:
        for variable in df.columns:
            # La file del pais correspondiente
            country_data = df[df["Country Name"] == key_country]
            
            #Esta línea de código es importante. Nos dice que, si (1) encuentra un NA en alguna de las variables
            #para (2) el país estudiado, entonces inicie el código.
            if not country_data.empty and country_data[variable].isna().values[0]:
                proxy_list = []
                #Lo que hace el código es agrupar los valores de esta variable para sus países vecinos
                #La idea es luego tomar el promedio de esta lista.
                for neighbour in neighbours_dict.get(key_country):
                    neighbour_data = df[df["Country Name"] == neighbour]
                    if not neighbour_data.empty and not neighbour_data[variable].isna().values[0]:
                        value = neighbour_data[variable].values[0]
                        proxy_list.append(value)
                
                # Solo calculamos el promedio si hay valores en la lista
                if len(proxy_list) > 0:
                    proxy_val = statistics.mean(proxy_list)
                    df.loc[df["Country Name"] == key_country, variable] = df.loc[df["Country Name"] == key_country, variable].fillna(proxy_val)
```

Un problema habitual del modelo _K-Means_ es que pierde desempeño a mayor dimensionalidad.

En esta pregunta defina **por lo menos 5 combinaciones de columnas** (mínimo 2 columnas y máximo 4). Para cada combinación determine el número óptimo de clusters mediante el "Elbow method" considerando $k \in [1, 30]$.

Un posible enfoque para lo anterior es tomar un _approach_ estadístico y escoger variables que empíricamente muestran mayores indicios de poder diferenciador entre las observaciones. De todas formas, recuerde que la pregunta de interés es "¿qué países se verán expuestos de forma similar a la guerra comercial?". En este sentido, es altamente recomendable escoger variables con una mayor relación al fenómeno de la guerra comercial, ya que puede ser útil para discernir de qué forma podría verse expuesto cada grupo.


```python
df_scaled.columns
```

```python
def elbow(df):
    dispersion = []
    for k in range(1, 31):
      model = KMeans(n_clusters=k, n_init='auto')
      model.fit(df)
      #Distancia euclidiana para cada par (dato - centro del cluster)
      dist_ecl = cdist(df, model.cluster_centers_, 'euclidean')
      #Tomamos la distancia mínima (al centroide más cercano)
      min_dist = np.min(dist_ecl, axis=1)
      #Promediamos: sumamos todas las mínimas distincia y dividimos por la cantidad de datos
      average_dis = sum(min_dist)/len(df)
      dispersion.append(average_dis)
    return dispersion
```

```python
# Grupo 1: Fuel exports, Manufactures Exports, Agricultural raw materials exports
grupo1 = ["Fuel exports (% of merchandise exports)", "Manufactures exports (% of merchandise exports)",
          "Agricultural raw materials exports (% of merchandise exports)"]
dispersion = elbow(df_scaled[grupo1])

fig = plt.figure(figsize=(15, 5))
plt.plot(range(1, 31), dispersion)
plt.grid(True)
plt.title('Elbow curve')
```
